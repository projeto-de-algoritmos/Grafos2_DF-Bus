```python
from queue import PriorityQueue
import json
```

```python
with open('cities.json', encoding='utf-8') as json_file:
    adjacency_list = json.load(json_file)
```

```python
adjacency_list
```

```python
def dijkstra(source, destiny):
    pq = PriorityQueue()
    dist = dict()
    
    for i in adjacency_list:
        dist[i] = (999999, '')

    dist[source] = (0, source)
    pq.put((0, source))
    
    while not pq.empty():
        d, u = pq.get()
        for dest in adjacency_list[u]:
            w = adjacency_list[u][dest]
            if(dist[dest][0] > d + w):
                dist[dest] = (d + w, u)
                pq.put((dist[dest][0], dest))
                
    path = [] 
    d = destiny
    path.append(d)
    
    
    trip_fare = dist[d][0]
    print(trip_fare)
    
    while path[-1] != source:
        path.append(dist[d][1])
        d = dist[d][1]
    path.reverse()
    print(path)

dijkstra('Santa Maria', 'Sobradinho II')
```
